```python
numerics = numerics.drop(['id','time'], axis=1)
```

Before we apply PCA we need to standardize/scale the data. For this we can use sklearn's preprocessing library 


```python
!pip install scikit-learn
from sklearn.preprocessing import StandardScaler
transformer = StandardScaler().fit(numerics)
scaled = transformer.transform(numerics)
scaled = pd.DataFrame(scaled)
scaled.columns = numerics.columns 
scaled.head()
```

```python
from sklearn.decomposition import PCA
from sklearn import preprocessing
pca = PCA()
pca.fit(scaled)
```

```python
pca.components_
```

```python
# variations = pca.explained_variance_
variations = pca.explained_variance_ratio_*100
variations = np.round(variations, decimals=1)
variations
```

```python
plot_labels = ['PC'+str(i) for i in range(1,len(variations)+1)]
?plot_labels
```

```python
from matplotlib.pyplot import figure
figure(figsize=(20,5))
plt.bar(x = plot_labels, height=variations)
plt.xlabel('Principal Components')
plt.ylabel('Variance Explained')
plt.title('PCA Scree Plot')
```

```python
# Libraries
import math
import pandas as pd # manipulate dataframes
import numpy as np # numerical python
import matplotlib.pyplot as plt # viz
import scipy.stats as st # stats
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
miu = 120
sample_population = 100
sample_mean = 130.1
std_population = 21.21
```

```python
null_hypothesis = "The mean of the systolic blood pressure isn't higher than 120.0 mm Hg."
alt_hypothesis = "The mean of the systolic blood pressure is higher than 120.0 mm Hg."

print(f"Null hypothesis (H0): {null_hypothesis}")
print(f"Alternative hypothesis (H1): {alt_hypothesis}")
```

```python
#defining the alpha
alpha = 0.05
print(f"Level of significance: {alpha}")
```

The way I'm going to test these hypothesis is the z test.

The formula to complete this is the following:

(sample_mean - population_mean) / population std / square_root of sample size

translated into (130.1 - 120) / 21.21 / square root of 100

```python
z_stat = (sample_mean - miu) / (std_population / (sample_population ** 0.5))
print(f"Test statistic (z): {z_stat}")
```

Since the z test value is above 3 we can confirm with elevated confidence that the H0 can be rejected in favor of H1.
